```python
#Question1:
#Answer:
""" The reason why we use Exception class while creating a custom Exception
    1. Consistent Exception Handling
    2. Clear Hierarchical Structure
    3. Predefined Exception Properties
    4. Compatibility with System Exceptions"""
    
class  invalid_email(Exception):
    def __init__(self,msg):
        self.msg = msg
```

```python
def check_email(email):
    if "@" not in email or "." not in email:
        raise invalid_email("email address is not valid")
    else:
        print("valid email")
```

```python
try:
    email = input("enter your email")
    check_email(email)
except invalid_email as e:
    print(e)
```

```python
#Question3:
#Answer:
"""The ArithmeticError class in Python is the base class for exceptions that are raised for various arithmetic errors.
It includes several specific exception classes that represent specific arithmetic-related issues.
Two common exceptions derived from ArithmeticError are ZeroDivisionError and OverflowError."""

try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(e)
    
```

```python
import sys

try:
    big_number = sys.maxsize
    result = big_number * big_number
except OverflowError as e:
    print(e)
```

```python
#Question4:
#Answer:
"""The LookupError class in Python is the base class for exceptions that occur when a lookup or indexing operation is performed on a sequence (like lists, tuples, dictionaries, etc.) and the key or index is not found or invalid.
It's a convenient way to catch multiple lookup-related exceptions with a single except block.
Two common exceptions derived from LookupError are KeyError and IndexError."""

try:
    d = {"name":"aman","time":6}
    print(d["key2"])
except KeyError as e:
    print(e)
```

```python
try:
    l = [1,2,3,4]
    print(l[6])
except IndexError as e:
    print(e)
```

```python
#Question5:
#Answer:
"""ImportError is a built-in exception class in Python that is raised when an import statement fails to locate and load a module or when there is an issue with importing a module."""
try:
    import aman
except ImportError as e:
    print(e)
```

```python
"""ModuleNotFoundError is a specific type of ImportError that is raised when the Python interpreter cannot find the specified module during the import process. It was introduced in Python 3.6 as a more informative and distinct exception for handling module-related import errors."""

try:
    import system123
except ModuleNotFoundError as e:
    print(e)
```
